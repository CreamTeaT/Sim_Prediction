```python
def nha_Mang(st):
  bac1 = ["086", "096", "097", "098", "032", "033", "034", "035", "036", "037", "038", "039"];
  bac2 = ["088", "091", "094", "083", "084", "085", "081", "082", "090", "093", "070", "079", "077", "076", "078"];
  bac3 = ["092", "056", "058", "099", "059", "087"];

  s = str(st)
  scheck = '0' + s[0] + s[1];
  if(scheck in bac1):
     return 4
  if(scheck in bac2):
     return 3
  if(scheck in bac3):
     return 2
  return 1

data["nha_mang"] = data["sim_number"].apply(nha_Mang)
data.head()
```

```python
def increaseLength_orderly(st):
  s = str(st)
  n = len(s);
  m = 1;

  for i in range(1, 9):
    if(ord(s[n-i-1]) + 1 == ord(s[n-i])):
      m += 1;
    else: break
  if(m >= 3): return m
  return 0

data["sim_tien_len"] = data["sim_number"].apply(increaseLength_orderly)
data.head()

for i in range (0, 199000):
  if data["sim_tien_len"][i] == 3 :
      print(data["sim_number"][i])
```

```python
#tiến lên đơn lẻ/chẵn
def tienlen_don_lechan(st):
  s = str(st)
  n = len(s);
  m = 1;

  for i in range(1, 9):
    if(ord(s[n-i-1]) + 2 == ord(s[n-i])):
      m += 1;
    else: break
  if(m >= 3): return m
  return 0

data["sim_tien_len_lechan"] = data["sim_number"].apply(tienlen_don_lechan)
data.head()

for i in range (0, 199000):
  if data["sim_tien_len_lechan"][i] == 3 :
      print(data["sim_number"][i])
```

GÁNH ĐẢO
* GÁNH ĐƠN: **aba, **abba
* Gánh kép bằng : **abcab
* Gánh kép tiến : **abcba

```python
# Gánh đơn
# * Ở đuôi
def check_sub_Palindrome(st):
  s = str(st)
  n = len(s)
  for i in range (n-1,0,-1):
    if(s[i-1]==s[i-2] and s[i] == s[i-3] and s[i] != s[i-1]): return 1
    elif(s[i] == s[i-2] and s[i] != s[i-1]): return 0
    else: break
  return 0

# *Ở giữa
def check_sub_Palindrome_mid(st):
  s = str(st)
  n = len(s)
  for i in range (n-2,2,-1):
    if(s[i-1]==s[i-2] and s[i] == s[i-3]and s[i] != s[i-1]): return 1
    elif(s[i] == s[i-2]and s[i] != s[i-1]): return 1
  return 0
data["Sim_Ganh_Dao_Duoi"] = data["sim_number"].apply(check_sub_Palindrome)
data["Sim_Ganh_Dao_Giua"] = data["sim_number"].apply(check_sub_Palindrome_mid)
data.head()

for i in range (0, 199000):
  if data["Sim_Ganh_Dao_Duoi"][i] == 1 :
      print(data["sim_number"][i])
```

```python
# lặp 4 12121212
def lap4(st):
  if(checkDuplicate(st) == 2):
    s = str(st)
    n = len(s)
 
    sc1 = s[1:3]
    sc2 = s[3:5]

    if(sc1 == sc2): return 1
    return 0
  return 0
data["Sim_Lap_Kep_1_1"] = data["sim_number"].apply(lap4)
data.head()

for i in range (0, 199000):
  if data["Sim_Lap_Kep_1_1"][i] == 1 :
      print(data["sim_number"][i])
```
