## Sim tiến lên

```python
def increaseLength_orderly(st):
  s = str(st)
  len = 1
  max = 1
  for i in range(0,8):
    if(ord(s[i]) + 1 == ord(s[i+1])):
      len += 1
      if(max < len) : max = len
    else:
      len = 1
  if(max < 3): max = 0
  return max
  
data["sim_tien_len"] = data["sim_number"].apply(increaseLength_orderly)
test["sim_tien_len"] = test["sim_number"].apply(increaseLength_orderly)
test.head()
```

## Sim lùi

```python
def decreaseLength_orderly(st):
  s = str(st)
  len = 1
  max = 1
  for i in range(0,8):
    if(ord(s[i]) - 1 == ord(s[i+1])):
      len += 1
    else:
      if(max < len) : max = len
      len = 1
  if(max < len) : max = len
  return max if max > 2 else 0

data["sim_lui"] = data["sim_number"].apply(decreaseLength_orderly)
test["sim_lui"] = test["sim_number"].apply(increaseLength_orderly)
test.head()
```

```python
def check9(st):
  s = str(st)
  n = len(s)
  for i in range(1, n) :
      if s[i] != s[0] :
          return 0
  return 1
data["sim_cuu_quy"] = data["sim_number"].apply(check9)
test["sim_cuu_quy"] = test["sim_number"].apply(increaseLength_orderly)
test.head()
```

## Dạng1: abab^n

```python
# Ở đuôi
def checkDuplicate(st):
  s = str(st)
  n = len(s)
  temp = [s[i-2:i] for i in range(n,1,-2)]
  count = 1
  for i in range (len(temp)-1):
    if(temp[i] == temp[i+1]): count +=1
    else: break
  if (count >= 1):
    return count  
  else:
    return 0
data["Sim_Lap_Kep_abab"] = data["sim_number"].apply(checkDuplicate)
test["Sim_Lap_Kep_abab"] = test["sim_number"].apply(checkDuplicate)
data.head()
```

```python
def checkDuplicate2(st):
  s = str(st)
  n = len(s)
  count = 1
  check = 0
  tinh = 0
  dem = 0
  for i in range (n-1):
    if(s[i] == s[i+1]):
      count += 1
      if(count > 2 and check == 0):
        check = 1
        tinh += 1
      if(count == 2): dem += 1
    else:
      count = 1
      check = 0
  dem = dem - tinh
  if(dem > 1):
    return dem
  else:
    return 0

data["Sim_Lap_Kep_aa"] = data["sim_number"].apply(checkDuplicate2)
test["Sim_Lap_Kep_aa"] = test["sim_number"].apply(checkDuplicate2)
data.head()
```
